## Model selection by cross-validation

Let's train different classifiers and optimize the one with the best cross-validation accuracy score. We train using the full dataset

```python
models = []

models.append(('lr_clf',  LogisticRegression(multi_class = 'multinomial', solver = 'newton-cg')))
models.append(('rf_clf',  RandomForestClassifier(random_state = 42)))
#models.append(('svm_clf',  SVC()))

cv_score = []
names = []

for name, model in models:
    
    score = cross_val_score(model, X_train, y_train, cv = 3, scoring = 'accuracy')
    cv_score.append(score)
    names.append(name)
    
    print('%s: %f(%f)'%(name, score.mean(), score.std()))
    print('****************************')  
```

##  Logistic regression:  hyperparameter tuning
